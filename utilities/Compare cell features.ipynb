## Setup

```python
# Import plotting packages
import matplotlib.pyplot as plt
import seaborn as sns

# Specify that all plots will happen inline & in high resolution
%matplotlib inline  
%config InlineBackend.figure_format = 'retina'

# Import pandas and numpy for working with databases
import pandas as pd
import numpy as np

print('Packages imported.')
```

## Load the dataset

```python
# Load 'cell_metrics.csv' which was generated previously
dataset = pd.read_csv('cell_metrics.csv')
dataset.head()
```

## Compare all of the features in this dataset, colored by dendrite type

Below, we'll use a really handy tool from Seaborn called [pairplot] which will plot a grid of all of our metrics against eachother. You can uncomment the line that says `set(xscale='log')` if you'd like to see what this looks like on a log scale.

```python
pairplot = sns.pairplot(dataset,hue='dendrite_type',height=2)
#pairplot.set(xscale='log',yscale='log')
plt.show()
```

# Choose metrics to compare
The pairplot above gives a great overview, but you may want to focus in on one of these interesting comparisons. To do so, we can use [scatterplot](https://seaborn.pydata.org/generated/seaborn.scatterplot.html).

```python
fig = plt.figure(figsize=(5,5))

scatter = sns.scatterplot(data= dataset,x = 'cell_surface_area', y = 'input_resistance', hue = 'dendrite_type')
scatter.set(xscale='log',yscale='log')
plt.show()
```
